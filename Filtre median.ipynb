```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import random
from scipy import ndimage
```

```python
orig_img = cv2.imread('/content/ronaldo.jpg')
```

```python
plt.title('Original image')
plt.imshow(orig_img)
```

```python
orig_img.shape
```

Adding salt&pepper noise

```python
def sp_noise(image,prob):

    output = np.zeros(image.shape,np.uint8)
    thres = 1 - prob
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            rdn = random.random()
            if rdn < prob:
                output[i][j] = 0
            elif rdn > thres:
                output[i][j] = 255
            else:
                output[i][j] = image[i][j]
    return output
```

```python
noised_img = sp_noise(orig_img,0.2)
```

```python
plt.title('Image denoised using median filter')
plt.imshow(noised_img)
```

denoising

```python
denoised_img= ndimage.median_filter(noised_img, size=10)
```
